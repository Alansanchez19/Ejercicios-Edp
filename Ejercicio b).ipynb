# Análisis de problema de Dirichlet (b)

La solución de la ecuación 
$$\frac{\partial^2 u}{\partial x^2}+ \frac{\partial^2 u}{\partial y^2}= 0,\;\;\;0<x<1,\;\;\;0<y<1$$
sujeta a las condiciones de frontera:
$$u(0,y)=10y, \;\frac{\partial u}{\partial x}=-1$$
Y la condición inicial:
$$ \;u(x,0)=u(x,1)=0$$

Esta dada en términos de la serie de Fourier por la expresión:
$$u(x,y)=\sum_{n=1}^{\infty} sin(n \pi y)[a_n cosh (n\pi y)+ b_n sin (n\pi y)].$$
Donde $a_n$ y $b_n$ son:
$$a_n=\frac{-20(-1)^n}{n\pi}$$

$$b_n=\frac{2[(-1)^n -1]+20(-1)^n (n\pi)sinh (n\pi)}{(n\pi)^2 cosh (n\pi)}$$

```python
from pylab import *
```

```python
x = linspace(0,1.0,1000)
y = linspace(0,1.0,1000)
```

```python
def u_x_y(x,y):
    N = 8
    f = zeros ((len(x),len(y)))
    for i in range(len(x)):
        for n in range(1,N):
          a = (-20*(-1)**n)/(n*pi)
          b = (2*((-1)**n-1)+20*(n*pi)*sinh(n*pi)*(-1)**n)/(n**2*pi**2*cosh(n*pi))
          f[:,i] += sin(n*pi*y)*(a*cosh(n*pi*x[i])+b*sinh(n*pi*x[i]))
    return f
```

Y podemos ver esto en una gráfica en 3D.

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection="3d")

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel('$y$'); plt.ylabel('$x$'); 
plt.title("u(x,y)")
ax.plot_surface(X, Y, U, cmap="inferno"); #Otros colores:"viridis", "inferno", "plasma"
```
